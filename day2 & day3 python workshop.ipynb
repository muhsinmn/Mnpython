### DROPPING

```python
loan_data.dropna(subset = ['Gender', 'Married'], inplace = True)
```

```python
loan_data['Gender'].isnull().sum()
```

```python
loan_data['Married'].isnull().sum()
```

```python
loan_data.shape
```

```python
loan_data['Dependents'].fillna(method = 'bfill',inplace = True)
```

```python
loan_data['Dependents'].isnull().sum()
```

```python
loan_data['Credit_History'].fillna(method = 'bfill',inplace = True)
#bfill mean they used the value that is below
```

```python
loan_data['Credit_History'].isnull().sum()
```

```python
loan_data.isnull().sum()
```

```python
loan_data['Credit_History'].fillna(method = 'pad',inplace = True)
# pad is filling by taking the above
```

```python
loan_data['Credit_History'].isnull().sum()
```

```python
loan_data['Self_Employed'].fillna(loan_data['Self_Employed'].mode()[0], inplace = True)
# the maximum number is used to fill
# it is used for characterical value
```

```python
loan_data['Self_Employed'].isnull().sum()
```

```python
loan_data['LoanAmount'].fillna(np.mean(loan_data['LoanAmount']), inplace = True)
# mean is used for numerical value
```

```python
loan_data['LoanAmount'].isnull().sum()
```

```python
loan_data.isnull().sum()
```
