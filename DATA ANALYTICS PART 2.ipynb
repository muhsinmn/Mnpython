#### Dealing with NULL values in the dataset

##### Drop with null values

```python
loan_data.isnull().sum()
```

```python
loan_data['Married'].dropna(axis = 0, inplace = True)
```

```python
loan_data['Married'].isnull().sum()
```

```python
loan_data.shape
```

```python
loan_data.dropna(subset = ['Married','Gender'], inplace = True)
```

```python
loan_data['Married'].isnull().sum()
```

```python
loan_data['Dependents'].fillna(method = 'bfill', inplace = True)
```

```python
loan_data['Dependents'].isnull().sum()
```

```python
loan_data['Credit_History'].fillna(method = 'pad', inplace = True)
```

```python
loan_data['Credit_History'].isnull().sum()
```

```python
loan_data['Self_Employed'].fillna(loan_data['Self_Employed'].mode()[0], inplace = True)
```

```python
loan_data['Self_Employed'].isnull().sum()
```

```python
loan_data['LoanAmount'].fillna(np.mean(loan_data['LoanAmount']), inplace = True)
```

```python
loan_data['LoanAmount'].isnull().sum()
```

```python
loan_data['Loan_Amount_Term'].fillna(np.mean(loan_data['Loan_Amount_Term']), inplace = True)
```

```python
loan_data['Loan_Amount_Term'].isnull().sum()
```
